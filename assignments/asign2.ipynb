```python
from scipy.stats import zscore

z_scores = zscore(data[['FY CGPA', 'SY CGPA']])
# print(z_scores)
data_no_outliers = data[(z_scores < 3).all(axis=1)] 
data_no_outliers
```

```python
outliers_mask = (z_scores > 3) | (z_scores < -3)

# Create a new column to flag outliers
data['Outlier'] = outliers_mask.any(axis=1)
data[data['Outlier']]
```

```python
mean_values = data.mean()

for column in ['FY CGPA', 'SY CGPA']:
    data[column]
```

```python
# Change the Scale for Better Understanding:
# Scaling 'SY CGPA' using Min-Max scaling to a range of [0, 1]
min_value = data['SY CGPA'].min()
max_value = data['SY CGPA'].max()
data['SY CGPA_scaled'] = (data['SY CGPA'] - min_value) / (max_value - min_value)
```

```python
# Convert a Non-linear Relation into a Linear One:
# Applying a square root transformation to 'SY CGPA'
data['SY CGPA_sqrt'] = np.sqrt(data['SY CGPA'])
```

```python
# Decrease Skewness and Convert Distribution into a Normal Distribution:
# Applying a log transformation to 'SY CGPA' (assuming it's positive)
data['SY CGPA_log'] = np.log1p(data['SY CGPA'])
```

```python

# Plotting original and transformed distributions
sns.histplot(data['SY CGPA'], kde=True, label='Original')
sns.histplot(data['SY CGPA_scaled'], kde=True, label='Scaled')
sns.histplot(data['SY CGPA_sqrt'], kde=True, label='Square root')
sns.histplot(data['SY CGPA_log'], kde=True, label='Log-transformed')
plt.legend()
plt.show()
```
